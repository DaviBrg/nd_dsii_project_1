````python

def add_diff_column(df, column, diff_name, fill_value):
    diff = df[column] - df[column].shift()
    diff.fillna(fill_value, inplace=True)
    df[diff_name] = diff
    return df

def get_hourly_entries(df):
    return add_diff_column(df, 'ENTRIES', 'ENTRIESn_hourly', 1)

turnstile_data = get_hourly_entries(turnstile_data)
turnstile_data.head()
````

### *Exercicio 2.4

Calcule agora a média de 'meantempi' nos dias que são sábado ou domingo (finais de semana):

````python
from datetime import datetime

def get_weekday_from_date_string(date_string):
    return datetime.strptime(date_string,'%Y-%m-%d').weekday()

def add_weekday_column(df):
    df['Weekday'] = df['DATEn'].apply(get_weekday_from_date_string)
    
def avg_weekend_temperature(filename):
    df = pd.read_csv(filename)
    add_weekday_column(df)
    mean_min_temp = df.groupby('Weekday').mean()['meantempi']
    mean_min_temp_weekend = mean_min_temp.loc[[5,6]]
    mean_min_temp_weekend.rename(index={5:"Média da mínima no sábado",6:"Média da mínima no domingo"}, inplace=True)
    return mean_min_temp_weekend

avg_weekend_temperature(filename)
````

### *Exercicio 2.5

Calcule a média da temperatura mínima 'mintempi' nos dias chuvosos onde da temperatura mínima foi maior que do 55 graus:

````python
def avg_min_temperature(filename):
    df = pd.read_csv(filename)
    df['rain'] = df['rain'].apply(int)
    df = df[df['mintempi'] > 55]
    avg_min_temp_rainy = df.groupby('rain').mean()['mintempi']
    avg_min_temp_rainy.rename(index={0:'Não Choveu',1:'Choveu'},inplace=True)
    return avg_min_temp_rainy['Choveu']

avg_min_temperature(filename)
````

### *Exercicio 3.2*

Agora crie o redutor. Dado o resultado do mapeador do exercicio anterior, o redutor deve imprimir(Não retornar) uma linha por UNIT, juntamente com o número total de ENTRIESn_hourly.Ao longo de maio (que é a duração dos nossos dados), separados por uma guia. Um exemplo de linha de saída do redutor pode ser assim: 'R001 \ t500625.0'

Você pode assumir que a entrada para o redutor está ordenada de tal forma que todas as linhas correspondentes a uma unidade particular são agrupados.

````python
#!/usr/bin/python

import sys

def reducer():

    total_entries = 0
    old_key = None

    for line in sys.stdin:

        data = line.strip().split('\t')

        if len(data) != 2:
            continue

        current_key, entries = data

        if old_key and current_key != old_key:
            print("{0}\t{1}".format(old_key,total_entries))
            total_entries = 0

        old_key = current_key
        total_entries += float(entries)

reducer()
````
